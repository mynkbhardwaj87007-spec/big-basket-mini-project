# Step 8: Find out the outliers from the dataset according to the columns

```python
def handle_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"{column} = Allowed range: {lower_bound:.2f} to {upper_bound:.2f}")

    mean_value = df[column].mean()
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound),
                          mean_value,
                          df[column])

handle_outliers('sale_price')
handle_outliers('market_price')
print("Outliers handled successfully using IQR method.")
```

# **step 9 : plots or visuals**

```python
plt.figure(figsize=(3,2))
df['category'].value_counts().plot(kind='barh', color='#b5db2a')
plt.title("Number of Products by Category")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.show()
```

```python
plt.figure(figsize=(4,2))
df['brand'].value_counts().head(10).plot(kind='bar', color='#b5db2a')
plt.title("Top 10 Brands with Most Products")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.histplot(df['sale_price'], bins=50, kde=True, color='#b5db2a')
plt.title("Distribution of Sale Price")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(3,2))
sns.histplot(df['rating'].dropna(), bins=20, color='#b5db2a')
plt.title("Distribution of Product Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# 5. Average Discount by Category
plt.figure(figsize=(3,2))
avg_discount = df.groupby('category')['discount'].mean().sort_values(ascending=False)
avg_discount.plot(kind='bar', color='#b5db2a')
plt.title("Average Discount by Category")
plt.ylabel("Average Discount ()")
plt.show()
```

```python
category_counts = df['category'].value_counts()

top6 = category_counts[:6]
others = category_counts[6:].sum()

top6['Others'] = others

plt.figure(figsize=(4,3))
top6.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='tab20')
plt.title("Category-wise Share of Products (Top 6 + Others)")
plt.ylabel("")
plt.show()
```
